```python
import coremltools as ct
import numpy as np
from PIL import Image
import cv2
import os
import matplotlib.pyplot as plt
```

```python
os.environ["OPENCV_IO_ENABLE_OPENEXR"]="1"
model = ct.models.MLModel('model.mlmodel')
```

```python
path = 'path-to-depth-map-image.exr'
img = cv2.imread(path)[:,:,0]*10 #convert to decimeters (can be done when creating coreml model too)
pil_image=Image.fromarray(img)
plt.imshow(pil_image)

print(f"image is size {img.shape}")
```

```python
out_dict = model.predict({'input_5': pil_image})
print(out_dict)
```
